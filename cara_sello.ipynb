```python
import random


class Player:
    def __init__(self, name):
        self.name = name
        self.points = 0

    def add_points(self):
        self.points += 1

def save_game_data(players):

    with open("game_data.txt", "w") as file:
        for player in players:
            file.write(f"{player.name},{player.points}\n")


def main():
    print("Bienvenido al juego de Cara o Sello")
    player_1_name = input("Ingrese el nombre del primer jugador: ")
    player_2_name = input("Ingrese el nombre del segundo jugador: ")
    player_1 = Player(player_1_name)
    player_2 = Player(player_2_name)

    print(f"{player_1.name} vs {player_2.name}")
    round = 1
    while True:
        current_player = random.randint(1, 2)
        print("1. Cara\n2. Sello")
        if current_player == 1:
            player_1_choice = int(input(f"{player_1.name}, ¿cara o sello?: "))
            if player_1_choice == 1:
                player_2_choice = 2
            else:
                player_2_choice = 1
        else:
            player_2_choice = int(input(f"{player_2.name}, ¿cara o sello?: "))
            if player_2_choice == 1:
                player_1_choice = 2
            else:
                player_1_choice = 1

        coin = random.randint(1, 2)
        if coin == 1:
            print("¡Cara!")
        else:
            print("¡Sello!")

        if player_1_choice == coin:
            player_1.add_points()
        if player_2_choice == coin:
            player_2.add_points()

        print(f"Puntos:\n{player_1.name} --> {player_1.points}\n{player_2.name} --> {player_2.points}")
        flag = input("Desea seguir jugando? (s/n): ")

        if flag == 'n':
            break

    # Save scores before exiting
    players = [player_1, player_2]
    save_game_data(players)


if __name__ == '__main__':
    main()
```
